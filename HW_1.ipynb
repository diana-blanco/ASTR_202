## 3. Planck Curve Numbers
### 3.1 Part a
Write an original code to integrate the Planck curve to find the total brightness B(T).

Planck Law: $B_\nu(T) = \frac{2h\nu^3/c^2}{exp(h\nu/kT)-1}$

```python
import matplotlib.pyplot as plt
import numpy as np
from   scipy.constants import c, h, k 
import scipy.integrate as integrate

temps = [5778] # randomly selected temperatures: Earth, Sun, hottest stars
nu = np.linspace(10**4, 10**16, 1000) # frequency range, from radio waves to gamma rays... 
for T in temps: 
    B = 2*h*nu**3/c**2/((np.exp(h*nu/(k*T))-1))
    total_B = scipy.integrate.trapz(B, nu)
    print("Total B(T) at T = %g K: %g W/m^2" % (T, total_B))
    plt.plot(nu, B, label = "T = %g K" % T)
    
plt.xlabel("Frequency [Hz]")
plt.ylabel("$B(T)$")
plt.legend()
```

### 3.2 Part b
Write an original code to find the numerical value of the peak of the Planck curve in terms of multiples of $x = h\nu/kT$. Repeat for the planck curve $B_{\lambda}(\lambda, T)$ and the variable $y = hc/\lambda kT$. 

For the first part of the problem, we express Planck's Law as a function of dimensionless variable x: 

$B_{\nu}(x,T) = \frac{2k^4T^4}{h^3c^2} \frac{x^3}{exp(x)-1}$

```python
from scipy import signal 

x = np.linspace(0, 10, 1000) # x range 

dless_B_x = x**3/(np.exp(x)-1)

# find peaks
peak = signal.find_peaks(dless_B_x)[0][0]
peak_x = x[peak]

plt.plot(x, dless_B_x, label = r"$x = \frac{h\nu}{kT}$")
plt.scatter(peak_x, dless_B_x[peak], s = 10)
plt.annotate(r"Peak of %g at $x$ = %g" % (dless_B_x[peak], peak_x), (peak_x, dless_B_x[peak]))
    
plt.xlabel("x")
plt.ylabel("f(x), Dimensionless $B_{\nu}(x, T)$")
plt.legend()
plt.tight_layout()
```

We do the same for the second part of the problem, except we express Planck's Law as a function of the dimensionless variable y: 

$B_{\lambda}(y,T) = \frac{2k^5T^5}{c^3h^4}\frac{y^5}{e^y-1}$

```python
y = np.linspace(0, 10, 1000) # y range 

dless_B_y = y**5/(np.exp(y)-1)

# find peaks
peak = signal.find_peaks(dless_B_y)[0][0]
peak_y = y[peak]

plt.plot(y, dless_B_y, label = r"$y = \frac{hc}{\lambda kT}$")
plt.scatter(peak_y, dless_B_y[peak], s = 10)
plt.annotate(r"Peak of %g at $y$ = %g" % (dless_B_y[peak], peak_y), (peak_y, dless_B_y[peak]))
    
plt.xlabel("y")
plt.ylabel(r"f(y), Dimensionless $B_{\lambda}(y, T)$")
plt.legend(loc = "upper left")
plt.tight_layout()
```

### 3.3 Part c
Write an original code to find the numerical value of the peak of the derivative of the Planck curve $\partial_{\nu}(\nu,T)/\partial T$ with respect to ? in terms of $x = h\nu/kT$. 

First, we calculate the derivative of Planck's Law in terms of x. The calculation is included in the written portion of this submission: 

$\partial B_{\nu}(x, T)/ \partial T = \frac{2k^3T^2}{c^2h^2} \frac{x^4e^x}{(e^x-1)^2}$

Then, we proceed to find the peak.

```python
x = np.linspace(0, 10, 1000) # x range 

d_dT_dless_B_x = x**4*np.exp(x)/((np.exp(x) - 1)**2)

# find peaks
peak = signal.find_peaks(d_dT_dless_B_x)[0][0]
peak_x = x[peak]

plt.plot(x, d_dT_dless_B_x, label = r"$y = \frac{hc}{\lambda kT}$")
plt.scatter(peak_x, d_dT_dless_B_x[peak], s = 10)
plt.annotate(r"Peak of %g at $x$ = %g" % (d_dT_dless_B_x[peak], peak_x), (peak_x, d_dT_dless_B_x[peak]))
    
plt.xlabel("x")
plt.ylabel(r"f(x), Dimensionless $B_{\nu}(x, T)$")
plt.legend(loc = "upper left")
plt.tight_layout()
```
