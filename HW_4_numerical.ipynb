### Peak of the Synchrotron Emission Spectrum

Write original code to determine the peak of the synchrotron emission spectrum. 

$F(x) \equiv x \int^{\infty}_x K_{5/3}(\xi)d\xi$

Plot $F(x)$ over the domain $ x \in [0,3]$ and indicate where the maximum occurs, annotating the plot with the value of x where $F(x)$ is a maximum. 

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from mpmath import *
import scipy.integrate as integrate

x_min = 10**(-3)     # lower limit ~ 0
x_max = 3    # upper limit 

# produce of an array of x values from [0,3]
x = np.linspace(x_min, x_max, int(10e3))

integral = []
F_x = []

for i in range(len(x)): 
    integral.append(quad(lambda x_prime: besselk(5/3, x_prime), [x[i], inf]));
    F_x.append(x[i]*integral[i]);
```

```python
plt.plot(x, F_x)

max_y = max(F_x)
max_x = x[F_x.index(max_y)]

plt.scatter(max_x, max_y, label = "Peak at x = %.3g" % max_x)
plt.xlabel(r"$x = \omega/\omega_c$");
plt.ylabel(r"$F(x)$");
plt.legend(); 
plt.title("Synchrotron Emission Spectrum");
```
